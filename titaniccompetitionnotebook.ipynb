```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

This reads the training set from the file and prints its first 5 rows

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
```

This reads the training set from the file

```python
#Put the survived of the training data in the y_train
y_train=train_data.iloc[:,1]

#Dropping features which will not affect the model or training
train_data=train_data.drop(["Cabin","Name","Ticket"],axis=1)
test_data=test_data.drop(["Cabin","Name","Ticket"],axis=1)
#Replacing the strings with numbers
train_data=train_data.replace('male',1)
train_data=train_data.replace('female',0)
train_data['Embarked']=train_data['Embarked'].replace('C',0)
train_data['Embarked']=train_data['Embarked'].replace('Q',1)
train_data['Embarked']=train_data['Embarked'].replace('S',2)
train_data['Embarked'].fillna(round(train_data['Embarked'].mean(),0),inplace=True)

test_data=test_data.replace('male',1)
test_data=test_data.replace('female',0)
test_data['Embarked']=test_data['Embarked'].replace('C',0)
test_data['Embarked']=test_data['Embarked'].replace('Q',1)
test_data['Embarked']=test_data['Embarked'].replace('S',2)
test_data['Embarked'].fillna(round(test_data['Embarked'].mean(),0),inplace=True)

#Filling any missing values such as NaNs in the Age and Fare fields
av_age=test_data[['Age']].mean()
av_ager=round(av_age.Age,2)
test_data["Age"].fillna(av_ager,inplace = True)


av_age_test=test_data[['Age']].mean()
av_age_testr=round(av_age_test.Age,2)
test_data["Age"].fillna(av_age_testr,inplace = True)

av_fare_train=train_data['Fare'].mean()
av_fare_trainr=round(av_fare_train,2)
train_data.fillna(av_fare_trainr,inplace=True)

av_fare=test_data[['Fare']].mean()
av_farer=round(av_fare.Fare,2)
print(av_fare)
test_data["Fare"].fillna(av_farer,inplace=True)


train_data.head()
test_data.head()
```

Preproccessing of the data by removing unwanted features and filling missing cells

Drop the cabin column because only 30% has values

```python

x_train=train_data.iloc[:,3:10]
x_train,x_validation,y_train,y_validation=train_test_split(x_train,y_train,test_size=0.25,random_state=0)
x_test=test_data.iloc[:,2:10]
x_train.head()
y_train.head()
x_test.head()
```

For degugging purposes

```python
model=GradientBoostingClassifier(n_estimators=125,max_depth=5,random_state=0,learning_rate=0.1)
model.fit(x_train,y_train)
validation_score=model.score(x_validation,y_validation)
print(validation_score)
predict=model.predict(x_test)
predicted_output=pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predict})
predicted_output.to_csv('my_submission.csv', index=False)
print('Done')
print(predicted_output)
```
